```python
# For the datset currently loaded, plot the obstacles extracted
# from the first laser reading against the true obstacles.

from slam_eval.slam_eval import obsticle_compare
from ppaml_car.lasers import plot_obstacles

cx, cy, ct = dataset.init_x, dataset.init_y, dataset.init_angle
lx, ly, lt = lasers.car_loc_to_laser_loc(cx, cy, ct, dataset.a, dataset.b)
true_obstacles = [(x, y, 0.37) for (x, y) in dataset.ground_obstacles]
true_obstacles = np.array(true_obstacles)
obs_lasers = None
for ts in xrange(len(dataset.timestamps)):
    if dataset.ts2sensor[ts] == 'laser':
        obs_lasers = np.array(dataset.ts2laser[ts])
        break

got_obstacles = ransac.extract_obstacles(lx, ly, lt, laser_angles, laser_max_range, obs_lasers)
got_lasers = fast.readings_for_obstacles(lx, ly, lt, laser_angles, laser_max_range, got_obstacles)
got_llik = norm_log_pdf_id_cov(obs_lasers, got_lasers, 30.0)
print "Got {} obstacles; llik={}".format(len(got_obstacles), got_llik)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))
lasers.plot_lasers(lx, ly, lt, laser_angles, laser_max_range, true_obstacles, obs_lasers, axs[0])
lasers.plot_lasers(lx, ly, lt, laser_angles, laser_max_range, got_obstacles, obs_lasers, axs[1])

n_extras, score = obsticle_compare(got_obstacles, true_obstacles)
print "{} ground obstacles; {} out obstacles".format(len(true_obstacles), len(got_obstacles))
print "Map error: {}".format(score)

plot_obstacles(true_obstacles, axs[2], color=(0, 1, 0, 0.6))
plot_obstacles(got_obstacles, axs[2], color=(1, 0, 0, 0.6))
axs[2].plot([-7, -7, 7, 7, -7], [-7, 7, 7, -7, -7], 'k')
axs[2].set_xlim(-8, 8)
axs[2].set_ylim(-8, 8)
```
