```python
%matplotlib inline

from numutil import logaddexp_many
from numutil import norm_log_pdf_id_cov
from ppaml_car import data
from ppaml_car import fast
from ppaml_car import lasers

import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = data.Dataset.read("2_bend")
```

```python
a = 0.299541
b = 0.0500507

obstacles = [(x, y, 0.37) for (x, y) in dataset.ground_obstacles]
obstacles = np.array(obstacles)
laser_angles = lasers.default_laser_angles()
laser_max_range = lasers.default_laser_max_range()
```

```python
def obs_model_llik(true_lasers, obs_lasers):
    return norm_log_pdf_id_cov(obs_lasers, true_lasers, 30)
```

```python
# Use this to tune the obstacle radius:
x, y, t = (-6.1097138622119642, -0.054890012646175573, 0.0022182922480532222)
lx, ly, lt = lasers.car_loc_to_laser_loc(x, y, t, a, b)
true_lasers = fast.readings_for_obstacles(lx, ly, lt, laser_angles, laser_max_range, obstacles)
plt.figure(figsize=(7, 7))
lasers.plot_lasers(lx, ly, lt, laser_angles, laser_max_range, obstacles, obs_lasers, plt.gca())
obs_lasers = np.array(dataset.ts2laser[11])
obs_model_llik(true_lasers, obs_lasers)
```

```python
def lw_for_pose(ts, x, y, t, plot=False):
    obs_lasers = np.array(dataset.ts2laser[ts])
    lx, ly, lt = lasers.car_loc_to_laser_loc(x, y, t, a, b)
    true_lasers = fast.readings_for_obstacles(lx, ly, lt, laser_angles, laser_max_range, obstacles)
    lw = obs_model_llik(true_lasers, obs_lasers)
    if plot:
        fig = plt.figure(figsize=(9,8))
        ax1 = fig.add_subplot(2, 2, 1)
        ax2 = fig.add_subplot(2, 2, 2)
        ax3 = fig.add_subplot(2, 1, 2)
        ax4 = ax3.twinx()
        lasers.plot_lasers(lx, ly, lt, laser_angles, laser_max_range, obstacles, obs_lasers, ax1)
        lasers.plot_lasers(lx, ly, lt, laser_angles, laser_max_range, obstacles, true_lasers, ax2)
        ax3.plot(obs_lasers - true_lasers)
        # ax4.plot(obs_lasers, 'r')
        ax4.plot(true_lasers, 'g')
    return lw
```

```python
# Closer:
pose = (-6.1097138622119642, -0.054890012646175573, 0.0022182922480532222)
lw_for_pose(11, *pose, plot=True)
```

```python
# Close:
pose = (-6.1080997298338362, -0.056532313908826841, -0.0032118177425493697)
lw_for_pose(11, *pose, plot=True)
```

```python
# Far:
pose = (-6.4097138622119642, -0.014890012646175573, 0.1022182922480532222)
lw_for_pose(11, *pose, plot=True)
```

```python
np.random.seed(666)

# Example of which particles survive:
center = (-6.1097138622119642, -0.054890012646175573, 0.0022182922480532222)
noise = np.random.normal(loc=0, scale=0.3, size=(10000, 3))
particles = noise + center
logweights = [lw_for_pose(*pose) for pose in particles]

# Normalize weights.
logweights = np.array(logweights)
logweights -= logaddexp_many(logweights)
weights = np.exp(logweights)
assert np.abs(np.sum(weights) - 1.0) < 1e-6

# Resample.
indices = np.random.choice(
    len(particles),
    size=len(particles),
    p=weights)
new_particles = []
for index in indices:
    new_particles.append(particles[index])
new_particles = np.array(new_particles)
print "{} particles survived".format(len(set(indices)))
    
def scatter_particles(particles, ax):
    ax.scatter(
        particles[:, 0],
        particles[:, 1],
        c=particles[:, 2],
        s=10,
        linewidths=0,
        vmin=-np.pi,
        vmax=np.pi)

# Plot before and after resampling.
fig, axs = plt.subplots(1, 2, figsize=(9, 4))

lasers.plot_obstacles(obstacles, axs[0])
scatter_particles(particles, axs[0])
axs[0].set_xlim(-8, 8)
axs[0].set_ylim(-8, 8)
axs[0].scatter([center[0]], [center[1]], marker='x')

lasers.plot_obstacles(obstacles, axs[1])
scatter_particles(new_particles, axs[1])
axs[1].set_xlim(-8, 8)
axs[1].set_ylim(-8, 8)
axs[1].scatter([center[0]], [center[1]], marker='x')
```

```python
def compare_particles(ts, *poses):
    obs_lasers = np.array(dataset.ts2laser[ts])
    fig, axs = plt.subplots(1, len(poses), figsize=(9, 4))
    lw_0 = None
    for i, (x, y, t) in enumerate(poses):
        lx, ly, lt = lasers.car_loc_to_laser_loc(x, y, t, a, b)
        true_lasers = fast.readings_for_obstacles(lx, ly, lt, laser_angles, laser_max_range, obstacles)
        lasers.plot_lasers(lx, ly, lt, laser_angles, laser_max_range, obstacles, obs_lasers, axs[i])
        lw = obs_model_llik(true_lasers, obs_lasers)
        if i == 0:
            lw_0 = lw
            xlikely = 1
        else:
            xlikely = np.exp(lw - lw_0)
        print "particle {} {}: lw={} ({}x as likely as particle 0)".format(
            i, (x, y, t), lw, xlikely)
```

```python
southernmost = min(new_particles, key=lambda p: p[1])
compare_particles(11, center, southernmost)
```

```python
# Use this to debug when the PF is making strange choices.
best_particle = [-1.7822291790904501, -5.2422048159613155, -1.9198023291529807]
ground_gps = [-0.689727, -6.24296, -1.5775]
compare_particles(9061, ground_gps, best_particle)
```
