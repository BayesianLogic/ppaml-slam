```python
%matplotlib inline

from numutil import logaddexp_many
from numutil import norm_log_pdf_id_cov
from ppaml_car import data
from ppaml_car import draw_dr
from ppaml_car import fast
from ppaml_car import lasers
from ppaml_car.pf import normalize_radians

import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = data.Dataset.read("2_bend")
```

```python
a = 0.299541
b = 0.0500507

obstacles = [(x, y, 0.37) for (x, y) in dataset.ground_obstacles]
obstacles = np.array(obstacles)
laser_angles = lasers.default_laser_angles()
laser_max_range = lasers.default_laser_max_range()
```

```python
plt.figure(figsize=(8,8))
ax = plt.gca()
gps_traj = draw_dr.get_ground_truth_traj(dataset)
ax.plot(gps_traj[:, 1], gps_traj[:, 2], label='ground')
dr_traj = draw_dr.get_dead_reckoning_traj(dataset, draw_dr.dynamics)
ax.plot(dr_traj[:, 1], dr_traj[:, 2], label='dead-reckoning')
ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
ax.legend()
```

```python
# What's the magnitude of the controls?
velocities = []
steerings = []
for ts in xrange(len(dataset.timestamps)):
    if dataset.ts2sensor[ts] == 'control':
        velocities.append(dataset.ts2control[ts][0])
        steerings.append(dataset.ts2control[ts][1])

fig, axs = plt.subplots(2, 1, figsize=(9,9))
axs[0].hist(velocities, bins=100)
axs[1].hist(steerings, bins=100)
None    
```

```python
# How much do x, y, theta actually change in one timestep?
dxs = gps_traj[1:,1] - gps_traj[:-1,1]
dys = gps_traj[1:,2] - gps_traj[:-1,2]
dthetas = gps_traj[1:,3] - gps_traj[:-1,3]
dthetas[dthetas < np.pi] += 2 * np.pi
dthetas[dthetas > np.pi] -= 2 * np.pi

fig, axs = plt.subplots(3, 1, figsize=(9,9))
axs[0].hist(dxs, bins=100)
axs[1].hist(dys, bins=100)
axs[2].hist(dthetas, bins=100)
None
```

```python
# Example trajectories if we add noise to pose, but not to controls:

def dynamics_with_pose_noise(dataset, old_state, velocity, steering, delta_t):
    new_state = draw_dr.dynamics(dataset, old_state, velocity, steering, delta_t)
    noise = np.random.normal(loc=0.0, scale=0.01, size=3)
    new_state[0] += noise[0]
    new_state[1] += noise[1]
    new_state[2] += noise[2]
    new_state[2] = normalize_radians(new_state[2])
    return new_state

plt.figure(figsize=(8,8))
ax = plt.gca()
ax.plot(gps_traj[:, 1], gps_traj[:, 2], label='ground')
ax.plot(dr_traj[:, 1], dr_traj[:, 2], label='dead-reckoning')
for i in xrange(5):
    dr_traj_noisy = draw_dr.get_dead_reckoning_traj(dataset, dynamics_with_pose_noise)
    ax.plot(dr_traj_noisy[:, 1], dr_traj_noisy[:, 2], label='noisy')
ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
ax.legend()
```

```python
# Example trajectories if we add noise to controls, but not to pose:

def dynamics_with_controls_noise(dataset, old_state, velocity, steering, delta_t):
    velocity += np.random.normal(loc=0.0, scale=0.1)
    steering += np.random.normal(loc=0.0, scale=0.3)
    return draw_dr.dynamics(dataset, old_state, velocity, steering, delta_t)

plt.figure(figsize=(8,8))
ax = plt.gca()
ax.plot(gps_traj[:, 1], gps_traj[:, 2], label='ground')
ax.plot(dr_traj[:, 1], dr_traj[:, 2], label='dead-reckoning')
for i in xrange(5):
    dr_traj_noisy = draw_dr.get_dead_reckoning_traj(dataset, dynamics_with_controls_noise)
    ax.plot(dr_traj_noisy[:, 1], dr_traj_noisy[:, 2], label='noisy')
ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
ax.legend()
```
